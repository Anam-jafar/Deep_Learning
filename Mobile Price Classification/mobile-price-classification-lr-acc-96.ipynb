```python
# Load Data

import pandas as pd
import numpy as np

dataset = pd.read_csv('/kaggle/input/mobile-price-classification/train.csv')
dataset.head()
```

```python
dataset.info()
```

# Training and Evaluation
Here, We will utilize various models in the dataset. The process will be as follows :
- Data Split to train and test
- Select model 
- Train on the selected model


```python
# Spliting data

from sklearn.model_selection import train_test_split

X = dataset.drop(columns= 'price_range', axis = 1)
Y = dataset['price_range']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 1)

print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model selection

from sklearn.linear_model import LogisticRegression

softmax_classifier = LogisticRegression(multi_class='multinomial', solver='newton-cg')
softmax_classifier.fit(X_train, Y_train)
```

```python
# Evaluation 
from sklearn.metrics import confusion_matrix, accuracy_score

Y_pred = softmax_classifier.predict(X_test)

acc = accuracy_score(Y_test, Y_pred)
print(acc)

print(confusion_matrix(Y_test, Y_pred))
```
