```python
import numpy as np
import pandas as pd
```

```python
# Dataset load
import sklearn.datasets 

data = sklearn.datasets.load_breast_cancer()
```

```python
data_frame = pd.DataFrame(data.data, columns = data.feature_names)
data_frame.head()
```

```python
# Add the labels to dataframe

data_frame['label'] = data.target
data_frame.head()
```

```python
# Find the value counts of each label

data_frame['label'].value_counts()
```

```python
# Analyze the values

data_frame.groupby('label').mean()
```

```python
# Feature Label Seperation and Data Split

from sklearn.model_selection import train_test_split

X = data_frame.drop(columns = 'label', axis = 1)
Y = data_frame['label']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, stratify = Y, test_size = 0.1, random_state = 2)
print(X_train.shape, X_test.shape, X.shape)
```

```python
# Model Selection

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
# Training

model.fit(X_train, Y_train)
```

```python
# Performance Evaluation

from sklearn.metrics import accuracy_score, confusion_matrix

Y_predicted = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_predicted)
cof_metrix = confusion_matrix(Y_test, Y_predicted)

print(f'Accuracy {accuracy}')
print(cof_metrix)
```
